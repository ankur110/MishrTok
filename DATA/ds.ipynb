```python
import os
import ast
import random
import unicodedata
from datasets import load_from_disk
from tqdm import tqdm
```

```python
OUTPUT_FILE = "shuffled_corpus.txt"
TARGET_HINDI_LINES = 5_000_000    # ~2-4 GB Hindi
TARGET_ENGLISH_LINES = 2_000_000  # ~1 GB English
HINGLISH_UPSAMPLE_FACTOR = 10 
```

```python
def clean_text(text):
    if not text or not isinstance(text, str):
        return None
    text = text.strip()
    if len(text) < 10:
        return None
    # NFKC for better compatibility (IndicSuperTokenizer style)
    return unicodedata.normalize('NFKC', text)
```

```python
hinglish_lines = []
```

```python
for row in tqdm(ds_hien_1["train"],desc="HINGLISH-1"):
    if(row.get("sentence2")):
        cleaned = clean_text(row["sentence2"])
        if cleaned:
            hinglish_lines.append(cleaned)
```

```python
for row in tqdm(ds_hien_2["train"],desc="HINGLISH-2"):
    if(row.get("Human-generated Hinglish")):
        raw=row["Human-generated Hinglish"]
        sentences=ast.literal_eval(raw)
        for sent in sentences:
            cleaned = clean_text(sent)
            if cleaned:
                hinglish_lines.append(cleaned)
    wac_text=row.get('WAC')
    if(wac_text):
        cleaned=clean_text(wac_text)
        if(cleaned):
            hinglish_lines.append(cleaned)
    pac_text=row.get('PAC')
    if(pac_text):
        cleaned=clean_text(pac_text)
        if(cleaned):
            hinglish_lines.append(cleaned)
    
```

```python
for row in tqdm(ds_hien_3["train"],desc="HINGLISH-1"):
    text=[]
    
    if(row.get('input')):
        text.append(row['input'].strip())
    if row.get('output'):
        text.append(row['output'].strip())
    if text:
        combined_text=" ".join(text)
        cleaned = clean_text(combined_text)
        if cleaned:
            hinglish_lines.append(cleaned)
```

```python
print(len(hinglish_lines))
```

```python
hinglish_lines = hinglish_lines * 10
random.shuffle(hinglish_lines)
```

```python
print(len(hinglish_lines))
```

```python
ds_hindi=ds_hi["train"].shuffle(seed=42)
```

```python
ds_english=ds_en["train"].shuffle(seed=42)
```

```python
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    for line in tqdm(hinglish_lines,desc="Hinglish"):
        f.write(line+"\n")
    count_hindi_lines=0
    for row in tqdm(ds_hindi,desc="Hindi"):
        cleaned=clean_text(row["text"])
        if(cleaned):
            f.write(cleaned+"\n")
            count_hindi_lines+=1
        if(count_hindi_lines>=TARGET_HINDI_LINES):
            break
    count_english_lines=0
    for row in tqdm(ds_english,desc="English"):
        cleaned=clean_text(row["text"])
        if(cleaned):
            f.write(cleaned+"\n")
            count_english_lines+=1
        if(count_english_lines>=TARGET_ENGLISH_LINES):
            break
```

```python
size_gb = os.path.getsize(OUTPUT_FILE) / (1024 ** 3)
total_approx = len(hinglish_lines) + count_hindi_lines + count_english_lines
print(f"\nDone! {OUTPUT_FILE} created ({size_gb:.2f} GB)")
print("Approximate proportions:")
print(f"  Hinglish: {len(hinglish_lines) / total_approx:.1%}")
print(f"  Hindi:    {count_hindi_lines / total_approx:.1%}")
print(f"  English:  {count_english_lines / total_approx:.1%}")
            
```
